```python
# ABISHEK NARAYANAN

# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading dataset
deliveries = pd.read_csv('deliveries.csv')
matches = pd.read_csv('matches.csv')

# Checking the shape of the dataset
print(deliveries.shape)

# Checking the columns in the dataset
print(deliveries.columns)

# Checking the first few rows of the dataset
print(deliveries.head())

# Checking the summary statistics of the dataset
print(deliveries.describe())

# Checking for missing values
print(deliveries.isnull().sum())

# Checking the shape of the dataset
print(matches.shape)

# Checking the columns in the dataset
print(matches.columns)

# Checking the first few rows of the dataset
print(matches.head())

# Checking the summary statistics of the dataset
print(matches.describe())

# Checking for missing values
print(matches.isnull().sum())

# Counting the number of matches won by each team
win_count = matches['winner'].value_counts()

# Pie-Chart percentage for Toss
toss_won = matches.toss_decision.value_counts()
labs1 = np.array(toss_won.index)
sizes = toss_won.values
colours = ['#FFA07A', '#20B2AA']  # Updated colors
plt.figure(figsize=(8,5))
plt.pie(sizes, labels=labs1, colors=colours, autopct=lambda pct: f"{pct:.1f}% ({int(pct * sum(sizes) / 100)})", shadow=True, startangle=90)
plt.title('Toss result', fontsize=18)
plt.axis('equal')
plt.show()

# Plotting Numbers of matches won by Toss result
plt.figure(figsize = (12,6))
sns.countplot(x='season',hue='toss_decision',data=matches,palette='hsv')
plt.title("Numbers of matches won by Toss result ")
plt.xlabel("Season",fontsize=10)
plt.ylabel("Count_Value",fontsize=10)
plt.show()

# Number of Matches played in each IPL season
plt.figure(figsize = (12,6))
sns.countplot(x='season',data=matches,palette="husl")
plt.title("Number of Matches played in each IPL season")
plt.xlabel("Season",fontsize=10)
plt.ylabel('Matches',fontsize=10)
plt.show()

# Pie-Chart percentage for Bat and Bowl First Win
matches['win_by'] = np.where(matches['win_by_runs'] > 0, 'Bat first', 'Bowl first')
bat_bowl = matches.win_by.value_counts()
labs = np.array(bat_bowl.index)
sizes = bat_bowl.values
colors = ['#008fd5', '#fc4f30']
plt.figure(figsize=(8, 5))
plt.pie(sizes, labels=labs, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90, textprops={'fontsize': 12, 'color': 'black'})
plt.title('Match Result', fontsize=18)
plt.axis('equal')
plt.show()

# Plot for Bat and Bowl first in Years
plt.figure(figsize=(12,6))
sns.countplot(x='season', hue='win_by', data=matches, palette='hsv')
plt.title("Numbers of matches won by batting and bowling first ")
plt.xlabel("Season", fontsize=10)
plt.ylabel("Count_Values", fontsize=10)
plt.show()

# Printing the most successful teams
most_successful_teams = win_count.index[:8].tolist()
print("Most Successful Teams:")
for teamss in most_successful_teams:
    print("- " + teamss)
# Plotting the number of matches won by each team
plt.figure(figsize=(12,6))
sns.barplot(x=win_count.index,y= win_count.values)
plt.xticks(rotation=50)
plt.title('Number of Matches Won by Each Team')
plt.ylabel('Number of Matches')
plt.xlabel('Team')
plt.show()

# Printing the Winner of every season
season_winners=matches.drop_duplicates(subset=['season'], keep='last')
print(season_winners[['season','winner']].reset_index(drop=True).sort_values('season'))

# Pie chart on Winning percentage in final
sw = season_winners.win_by.value_counts()
labs2=np.array(toss_won.index)
sizes = sw.values
colour2 = ['silver', 'red']
plt.figure(figsize = (8,5))
plt.pie(sizes, labels=labs2, colors=colour2,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Match Result',fontsize=18)
plt.axis('equal')
plt.show()

s_d=matches[['id','season','winner']]
comp_data=deliveries.merge(s_d,how='inner',left_on='match_id',right_on='id')
fours_d=comp_data[comp_data['batsman_runs']==4]
fours_d.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
batsman_fours=fours_d.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)

six_d=comp_data[comp_data['batsman_runs']==6]
six_d.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')])
batsman_six=six_d.groupby('batsman')['batsman_runs'].agg([('six','count')]).reset_index().sort_values('six',ascending=0)

# Printing the Numbers of fours hit by players
ax=batsman_fours.iloc[:10,:].plot('batsman','four',kind='bar',color='blue')
plt.title("Numbers of fours hit by players ")
plt.xticks(rotation=50)
plt.xlabel("Player name",fontsize=10)
plt.ylabel("No of fours",fontsize=10)
plt.show()

# Printing the Numbers of six hit by players
ax=batsman_six.iloc[:10,:].plot('batsman','six',kind='bar',color='red')
plt.title("Numbers of six hit by players")
plt.xticks(rotation=50)
plt.xlabel("Player name",fontsize=10)
plt.ylabel("No of six",fontsize=10)
plt.show()

# Counting the number of runs scored by each player
runs_count = deliveries.groupby('batsman')['batsman_runs'].sum().reset_index()

# Sorting the players in descending order of runs scored
runs_count = runs_count.sort_values('batsman_runs', ascending=False)

# Plotting the number of runs scored by each player
plt.figure(figsize=(12,6))
sns.barplot(x=runs_count['batsman'][:10], y=runs_count['batsman_runs'][:10], alpha=0.8)
plt.title('Number of Runs Scored by Each Player')
plt.ylabel('Number of Runs')
plt.xlabel('Player')
plt.show()

# Printing the most successful players
v=matches['player_of_match'].value_counts()[:10]
print("Top 10 most successful players:")
for player, count in v.items():
  print("- {} ({})".format(player, count))

# We can find the factors contributing to the win or loss of a team by performing a correlation analysis between 
# the various variables in the dataset.

# Creating a correlation matrix
corr_matrix = matches.corr()

# Plotting the correlation matrix
plt.figure(figsize=(12,6))
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()

# From the correlation matrix, we can see which variables have a strong positive or 
# negative correlation with the outcome variable, i.e., the result of the match.

# Convert the suggested teams and players into lists
endorsement_teams = [win_count.index[:6]][0].tolist()
endorsement_players = [v.index][0].tolist()

# Printing the suggested teams and players for endorsement
# Suggested teams for endorsement
print("Suggested teams for endorsement:")
for team in endorsement_teams:
    print("- " + team)

# Suggested players for endorsement
print("\nSuggested players for endorsement:")
for player in endorsement_players:
    print("- " + player)


```
